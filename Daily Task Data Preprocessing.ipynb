```python
import pandas as pd
import seaborn as sns
data = sns.load_dataset('titanic')
data.head()
```

```python
data = data.drop('who',axis=1)
data
data = data.drop('alive',axis=1)
data
data = data.drop('adult_male',axis=1)
data
data = data.drop('class',axis=1)
data
data = data.drop('embarked',axis=1)
data
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder,OneHotEncoder
```

```python
label_enc = LabelEncoder()
data['town_encoded'] = label_enc.fit_transform(data['embark_town'])
data
```

```python
sales_order = ['male','female']
ordinal_enc = OrdinalEncoder(categories=[sales_order])
data['sex_encoded'] = ordinal_enc.fit_transform(data[['sex']])
data
```

```python
deck_onehot = pd.get_dummies(data['deck'],prefix='deck')
deck_onehot
```

```python
deck_onehot = pd.get_dummies(data['deck'],prefix='deck',dtype=int,drop_first=True)
deck_onehot
```

```python
onehot = OneHotEncoder(sparse_output=False)
deck_onehot = onehot.fit_transform(data[['deck']])
deck_onehot
```

```python
onehot_df=pd.DataFrame(deck_onehot,columns=['A','B','C','D','E','F','G','NaN'])
onehot_df
```

```python
onehot_df=pd.DataFrame(deck_onehot,columns=onehot.get_feature_names_out(['deck']))
onehot_df
```

```python
new_data = pd.concat([data,onehot_df],axis=1)
new_data
```
